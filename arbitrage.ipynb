```python
dfl_1019 = dfl.loc[dfl.index.isin(pd.date_range("2015-10-19", "2015-10-20", freq="s"))]
dfb_1019 = dfb.loc[dfb.index.isin(pd.date_range("2015-10-19", "2015-10-20", freq="s"))]
```

```python
dfl_1019
```

```python
dfb_1019
```

```python
dfb_1019 - dfl_1019
```

```python
difference = (dfb_1019 - dfl_1019)
difference.describe()
```

```python
profit = difference[difference["Close"] > 0]
```

```python
profit
```

```python
profit_after_comission = (profit / dfl_1019) - 0.01
```

```python
profit_after_comission
```

```python
profit_after_comission = profit_after_comission.dropna()[profit_after_comission["Close"] > 0]
```

```python
profit_after_comission
```

```python
profit_after_comission["Close"].sum()
```

```python
profit_after_comission["cumsum"] = profit_after_comission["Close"].cumsum()
```

```python
profit_after_comission
```

```python
profit_after_comission.plot()
```

```python
profit_after_comission["cumprod"] = (profit_after_comission["Close"] + 1).cumprod()
```

```python
profit_after_comission
```

```python
invest = 1000
```

```python
# this is what you can make daily buying from LA and selling from Brazil including comission fees - typically done using pandas
profit_after_comission.iloc[-1]["cumprod"] * invest
```
